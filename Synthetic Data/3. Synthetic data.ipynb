```python
import pandas as pd
from datetime import datetime
import psycopg2

# ðŸ“Œ Load the CSV
df = pd.read_csv("synthetic_hr_emails_final.csv")

# âœ… Prepare records for insertion (ignore true_category column)
records = []
for idx, row in df.iterrows():
    email_id = int(row['email_id'])
    sender_email = row['sender_email']
    subject = row['subject']
    body = row['body']
    received_at = row['received_at']  # already in correct format

    records.append((email_id, sender_email, subject, body, received_at))
```

```python
# Your table schema:
# email_id | sender_email | subject | body | received_at

insert_query = """
INSERT INTO emails (email_id, sender_email, subject, body, received_at)
VALUES (%s, %s, %s, %s, %s)
"""

for record in records:
    cursor.execute(insert_query, record)

conn.commit()
print("âœ… All records inserted successfully.")
```
